```python
import numpy as np
import pandas as pd
```

```python
weather = 'assets/weather.csv'
```

```python
weather_data = pd.read_csv(weather)
```

```python
weather_data.head()
```

```python
weather_data.columns
```

```python
print weather_data.Depart.unique()
print weather_data.Tavg.unique()
print weather_data.PrecipTotal.unique()
print weather_data.AvgSpeed.unique()
print weather_data.Sunrise.unique()
print weather_data.Sunset.unique()
```

```python
weather_data['avg_temp'] = weather_data.Tavg.replace('M',np.nan)
weather_data['rainfall']=weather_data.PrecipTotal.replace(['  T','M'],'0.00')
weather_data['windspeed']=weather_data.AvgSpeed.replace('M', np.nan)
weather_data['departure']=weather_data.Depart.replace('M', np.nan)
weather_data['rise'] = weather_data.Sunrise.replace('-', np.nan)
weather_data['set'] = weather_data.Sunset.replace('-', np.nan)
```

```python
weather_data.rise.str[:2]
```

```python
weather_data['rise_hr'], weather_data['rise_min']= weather_data.rise.str[:2],weather_data.rise.str[2:]
weather_data['set_hr'], weather_data['set_min']= weather_data.set.str[:2],weather_data.set.str[2:]
```

```python
weather_data['hr_diff']=weather_data.set_hr.astype(float) - weather_data.rise_hr.astype(float)
weather_data['min_diff']=weather_data.set_min.astype(float) - weather_data.rise_min.astype(float)
weather_data['daylight_mins'] = weather_data.hr_diff*60 + weather_data.min_diff
```

```python
weather_data.daylight_mins.describe()
```

```python
weather_data.dropna(inplace=True)
weather_data.avg_temp = pd.to_numeric(weather_data.avg_temp)
weather_data.rainfall = pd.to_numeric(weather_data.rainfall)
weather_data.windspeed = pd.to_numeric(weather_data.windspeed)
weather_data.departure = pd.to_numeric(weather_data.departure)
weather_data.dtypes
```

```python
weather_data.windspeed.sum()
```

```python
all_weather = weather_data.groupby('Date').agg({'avg_temp':'mean', 'rainfall':'mean', 'windspeed':'mean', 'daylight_mins':'mean','departure':'mean'})
```

```python
all_weather.head(30)
```

```python
#all_weather = all_weather.sort_index
for var in ['avg_temp', 'rainfall', 'windspeed', 'daylight_mins','departure']:
    for i in range(1,15):
        all_weather[var + str(i) + "Days Before"] = all_weather[var].shift(i)
```

```python
all_weather['Last7Average_Temp'] = all_weather[all_weather.columns[5:12]].mean(axis=1)
all_weather['Last14Average_Temp'] = all_weather[all_weather.columns[5:19]].mean(axis=1)
all_weather['Last7Average_Windspeed'] = all_weather[all_weather.columns[33:40]].mean(axis=1)
all_weather['Last14Average_Windspeed'] = all_weather[all_weather.columns[33:47]].mean(axis=1)
all_weather['Last7Total_Rainfall'] = all_weather[all_weather.columns[19:26]].sum(axis=1)
all_weather['Last14Total_Rainfall'] = all_weather[all_weather.columns[19:33]].sum(axis=1)
all_weather['Last7Average_Daylight'] = all_weather[all_weather.columns[47:54]].mean(axis=1)
all_weather['Last14Average_Daylight'] = all_weather[all_weather.columns[47:61]].mean(axis=1)
all_weather['Last7Average_Departure'] = all_weather[all_weather.columns[61:68]].mean(axis=1)
all_weather['Last14Average_Departure'] = all_weather[all_weather.columns[61:75]].mean(axis=1)
```
